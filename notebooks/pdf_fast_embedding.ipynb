# Fast Embedding Exploration

This notebook explores fast embedding approaches for PDF documents.

```python
from rag_pipeline.config.parameter_sets import FAST_ANSWERS
from rag_pipeline.core.embeddings import process_pdf, query_embeddings

pdf = "tests/test_data/2303.18223v16.pdf"
persist = "data/fast_chroma"

chunks, records = process_pdf(
    pdf,
    FAST_ANSWERS.embedding.model_name,
    persist_dir=persist,
    chunk_size=FAST_ANSWERS.chunking.chunk_size,
    chunk_overlap=FAST_ANSWERS.chunking.chunk_overlap,
    max_pages=1,
)

print(chunks, records)
```

```python
for text in query_embeddings("transformer", FAST_ANSWERS.embedding.model_name, persist_dir=persist):
    print(text[:200])
```
