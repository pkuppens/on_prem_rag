# Context Rich Embedding Exploration

This notebook explores context-rich embeddings for PDF documents.

```python
from rag_pipeline.config.parameter_sets import CONTEXT_RICH
from rag_pipeline.core.embeddings import process_pdf, query_embeddings

pdf = "tests/test_data/2303.18223v16.pdf"
persist = "data/context_chroma"

chunks, records = process_pdf(
    pdf,
    CONTEXT_RICH.embedding.model_name,
    persist_dir=persist,
    chunk_size=CONTEXT_RICH.chunking.chunk_size,
    chunk_overlap=CONTEXT_RICH.chunking.chunk_overlap,
    max_pages=1,
)

print(chunks, records)
```

```python
for text in query_embeddings("transformer", CONTEXT_RICH.embedding.model_name, persist_dir=persist):
    print(text[:200])
```
